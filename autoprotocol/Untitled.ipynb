```python
import cv2
import numpy as np

# cv2.namedWindow("preview")
vc = cv2.VideoCapture(0)

if vc.isOpened(): # try to get the first frame
    rval, frame = vc.read()
else:
    rval = False

while rval:
#     cv2.imshow("preview", frame)
    rval, frame = vc.read()
    key = cv2.waitKey(20)
    avg = [0, 0, 0]
    readings_used = 0
    for row in range(len(frame)):
        for column in range(len(frame[row])):
            if row % 10 == 0 and column % 10 == 0:
                avg += frame[row][column]
                readings_used += 1
    print(avg / readings_used)
#     print(np.median(frame, axis=0))
# if key == 27: # exit on ESC
#     break
# cv2.destroyWindow("preview")
```
